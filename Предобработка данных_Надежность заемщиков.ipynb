На основании диапазонов, указанных ниже, создадим в датафрейме `data` столбец `total_income_category` с категориями:**

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.


```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
data['total_income_category'].value_counts() # Вот так?
```

```python
data['purpose'].unique()
```

Создадим функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:**

- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.


```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
data['purpose_category'].value_counts() 
```

# Исследовательский анализ 

###  Есть ли зависимость между количеством детей и возвратом кредита в срок?

```python
df = pd.DataFrame({'sum': [],
                    'count': [],
                    'mean': []
                  })

df['sum']=data.groupby(['children'])['debt'].sum()
df['count']=data.groupby(['children'])['debt'].count()
df['mean']=data.groupby(['children'])['debt'].mean()*100

df
```

**Вывод:** Да, зависимость есть. Показатель у семей с 5-ю детьми, возможно связан с маленьким количеством записей (это вполне логично, ведь семьи с 5-ю детьми встречаются гораздо реже). Наилучший показатель у людей, которые не имеют детей, доля должников здесь составляет с 7,5%, хуже обстоит дело с кредиторами, кто имеет одного или двух детей, здесь доля невозврата составляет соответственно 9,2% и 9,4%.

### Есть ли зависимость между семейным положением и возвратом кредита в срок?¶

```python
data.groupby(['family_status'])['debt'].agg(['count','mean'])

def categorize_family_status(status):
    try:
        if 'Не женат' in status or 'в разводе' in status:
            return 'не в браке'
        elif 'женат' in status or 'гражданский брак' in status:
            return 'в браке'
        elif 'вдовец' in status:
            return 'вдовец / вдова'
    except:
        pass
    
data['family_status_category'] = data['family_status'].apply(categorize_family_status)
data.groupby(['family_status_category'])['debt'].agg(['count','sum', 'mean'])
```

**Вывод:** Вдовцы/вдовы реже остальных имеют задолженности по кредитам. 

```python
# Разделим клиентов на группы по уровню зароботка, (Ранее мы получили 5 групп (E, D, C, B, A), 
#но сейчас я выделила всего четыре группы, так как у нас довольно мало записей для людей с низким доходом)
#D - клиент с наименьшим уровнем дохода
#A - клиент с наибольшим уровнем дохода

#print(data.groupby('total_income_category')['total_income'].sum())
def categorize_income(income):
    try:
        if income <= 200000:
            return 'Низкий доход'
        elif 200001 <= income:
            return 'Высокий доход'
        
    except:
        pass
data['total_income_category'] = data['total_income'].apply(categorize_income)
print(data.groupby(['total_income_category'])['debt'].agg(['count','mean']))

```

**Вывод:** Люди со средним доходом наиболее часто становятся должниками по кредиту, также в числе задолжников часто оказываются люди с высоким доходом. 

#Не совсем понимаю, как это объяснить, возможно, людям со средним доход тяжелее распределить бюджет, а люди с высоким доходом видимо берут кредиты для проведения крупных операций, что тоже может быть не так предсказуемо. А люди с низким доходом, я думаю, более экономны и заботятся о хорошей кредитной истории.

### Как разные цели кредита влияют на его возврат в срок?
