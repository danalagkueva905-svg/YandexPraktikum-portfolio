```python
data['price_sqm'] = data['last_price'] / data['total_area'] #цена за кв метр

def categorize_price(price):
    try:
        if 1.118349e+02 <= price <=  7.656613e+04:
            return 'D'
        elif  7.656614e+04 <= price <=  9.500000e+04:
            return 'C'
        elif  9.500001e+04 <= price <=   1.142132e+05:
            return 'B'
        elif   1.142133e+05 <= price <=  1.907500e+06:
            return 'A'
    except:
        pass
data['price_sqm_category'] = data['price_sqm'].apply(categorize_price)
#data['price_sqm_category'].value_counts()
data.groupby(['price_sqm_category'])['last_price'].agg(['count','sum', 'mean'])

#data.groupby(['price_sqm_category'])['price_sqm'].agg(['count','sum', 'mean'])

data['living_area'] = data['living_area'].fillna(data.groupby('price_sqm_category')['living_area'].transform('mean'))
#data['living_area'].isna().sum() #проверяем колич
```

```python
data['kith_proportion'] = data['kitchen_area'] / data['total_area'] *100  #цена за кв метр
data['kith_proportion'].describe()

def categorize_area(area):
    try:
        if 2.538071 <= area <=  13.333333:
            return 'маленькая кухня'
        elif  13.333334 <= area <=   17.236256:
            return 'средняя кухня'
        elif   17.236257 <= area <=   23.243243:
            return 'большая кухня'
        elif   23.243244 <= area <=  113:
            return 'очень большая кухня'
    except:
        pass

data['kitchen_area_category'] = data['kith_proportion'].apply(categorize_area)
#data['price_sqm_category'].value_counts()
data.groupby(['kitchen_area_category'])['kitchen_area'].agg(['count','sum', 'mean'])

data['kitchen_area'] = data['living_area'].fillna(data.groupby('kitchen_area_category')['kitchen_area'].transform('mean'))
data['kitchen_area'].isna().sum()
```

### Обработка дубликатов

**Что вызывает пробоему?**
* поселок/посёлок
* посёлок городского типа/село/городской посёлок,  городской посёлок/деревня - эти группы часто путают
* есть выбивающиеся типы населенных пунктов

1. заменим букву "е" на "ё" в слове "поселок"
2. населенные пункты, входящие в категрию "посёлок городского типа/село/городской посёлок" отнесем в одну грппу "посёлок"

### Посчитайте и добавьте в таблицу новые столбцы

```python
#день публикации объявления
data['day'] = data['first_day_exposition'].apply(lambda x: x.weekday())
weekdays_dict = {0: '0 - понедельник', 1: '1 - вторник', 2: '2 - среда', 
                 3: '3 - четверг', 4: '4 - пятница', 5: '5 - суббота', 6: '6 - воскресенье'}

data['day'] = data['day'].map(weekdays_dict)

#месяц публикации объявления
data['month'] = data['first_day_exposition'].dt.month

#год публикации объявления
data['year'] = data['first_day_exposition'].dt.year
```

```python
#тип этажа квартиры 

def categorize_floor(value):
    max_floor = value['floors_total']
    floor = value['floor']
    try:
        if floor == 1:
            return 'первый'
        elif  floor == max_floor:
            return 'последний'
        else:
            return 'другой'
    except:
        pass
    
data['floor_type'] = data.apply(categorize_floor, axis=1)
data['floor_type']
```

**Разобьем цены на три категории - недорогие, норальные и дорогие. Построим сводную таблицу для оценки средних значений.**

```python
def categorize_price(price):
    try:
        if price <=  3.400000e+06:
            return 'низкая цена'
        elif  3.400001e+06 <= price <= 6.790000e+06:
            return 'средняя цена'
        elif   6.790001e+06 <= price:
            return 'высокая цена'
    except:
        pass

data['last_price_category'] = data['last_price'].apply(categorize_price)
```

```python
stat = data.pivot_table(index='last_price_category', values=['total_area', 'living_area', 'rooms', 'ceiling_height', 'floor', 'cityCenters_nearest'])
print(stat)
```
